```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.integrate import quad
import scipy.special as sc
from scipy.integrate import odeint
import process_utility as psu
import scipy.integrate as integ
from scipy.optimize import curve_fit
import time
```

# SiPM2 at 32 V (threshold 15) - no ABALONE

```python
data = psu.read_file('/data/abalone/2021_03_26/SiPM2_t15.dat')
```

```python
psu.plot_waveforms2( data, 5)
```

```python
for i in range(5):
    plt.figure(figsize=(12,6))
    listpeaks = psu.search_peaks(data[i], 4, 2, False)
    psu.integral_central_peak(data[i],listpeaks, dtl=-8, dtr=1, tfit=8,
                              tlim=100,tc=5,tll=8,tlr=10,plot=1)
```

```python
nn = 10000
peakint = np.zeros(nn)
print('Total events:',len(data))
t_start = time.time()
for i in range(nn):
    listpeaks = psu.search_peaks(data[i], 4, 2, False)
    peakint[i] = psu.integral_central_peak(data[i],listpeaks,-8,1,10,100,5,8,10)
    diff = time.time() - t_start
    if (i % 1000) == 0:
        print(f'event n. {i} area: {peakint[i]:.2f}, time to process: {diff:.2f}')
```

```python
psu.histo_plot(peakint,1,20,500)
```
