```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.integrate import quad
import scipy.special as sc
from scipy.integrate import odeint
import scipy.integrate as integ
from scipy.optimize import curve_fit
import time
import json
```

```python
def read_waveforms_from_json(filename):
    wfs = []
    with open(filename) as f:
        for jsonObj in f:
            wfDict = json.loads(jsonObj)
            wfs.append(wfDict)
    plt.figure(figsize=(8,4.5))
    for i in range(len(wfs)-1):
        plt.plot(wfs[i]['voltages'],label=f'waveform {i}')
    plt.xlabel('time (ns)', ha='right', x=1)
    plt.ylabel('voltages (a.u.)', ha='right', y=1)
    plt.legend()
    plt.savefig('waveforms.png',dpi=600)
```
