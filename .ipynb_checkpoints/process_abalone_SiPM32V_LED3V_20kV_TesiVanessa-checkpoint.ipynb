```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from numpy import mean
```

```python
import scipy
from scipy.integrate import quad
import scipy.integrate as integ
from scipy.optimize import curve_fit
```

```python
data=pd.read_hdf('/home/dorazio/ABALONE/data_ABALONE_area11.h5')
```

```python
def gauss(x,a,mu,sigma):
    return a*np.exp(-(x-mu)**2 / (2.*sigma**2))
def landau(x,a,loc,scale):
    return 1.6*a*np.exp(-( (x-loc)/scale + np.exp(-(x-loc)/scale) )/2)
def expgauss(x,a,tau,x0):
    return a * 1/2 * np.exp((1+2*tau*x0-2*tau*x)/(2*tau**2))* scipy.special.erfc((1/tau+x0-x)/np.sqrt(2))
```

```python
def spe_spectrum_gauss_exp(x,a1,mu1,sigma1,a2,tau,x0):
    return gauss(x,a1,mu1,sigma1)+ expgauss(x,a2,tau,x0)

def fit_spe_spectrum_gauss_exp(area, bins = 200, volts = 10, ledv = 3, low = 0, high = 10000, spe_div = 2500, sig2=8, save = True):
    area_space = np.linspace(low,high, bins)
    h, t = np.histogram(area, bins=area_space) #h="altezza" della colonna dell'istogramma(sarebbe il valore sull'asse y) e t=estremo dell'intervallo corrispondente a h (sarebbe il valore sull'asse x) 
    plt.figure(figsize=(8,4.5))
    tc = (t[:-1] + t[1:])/2.  #tc= individua il punto centrale dell'intervallo della colonna dell'istogramma. In questo modo a ogni h corrisponde un solo tc
    a1 = plt.hist(area,bins=area_space,histtype='step',lw=2,density=False)
    
    idx1 = np.where(t>spe_div)[0][0]  #spe_div è il punto in cui decidiamo che deve iniziare il fit
    imax = np.argmax(h[idx1:])+idx1
    mu, hmax = t[imax], h[imax]
    idx = np.where(h[idx1:]>hmax/2) # fwhm=ampiezza a metà altezza
    ilo, ihi = idx[0][0], idx[0][-1]
    sig = (t[ihi]-t[ilo]) / 2.355
    
    guess = (hmax, mu, sig, 350, 2000, 400)#, hmax/2, 25, 5)
    
    bounds = ([hmax*0.9, mu-sig, 0, 300, 1900, 290],
              [1.1*hmax, mu+sig, 2*sig, 400, 2500, 450])
    #fit 
    
    
    popt, pcov = curve_fit(spe_spectrum_gauss_exp, tc[idx1:], h[idx1:], p0 = guess, bounds = bounds) #6 parametri ottimali che si ottengono con il fit: i primi 3 sono relativi alla gaussiana e gli ultimi 3 all'esponenziale convoluta con una gaussiana
    perr = np.sqrt(np.diag(pcov)) #stima degli errori
    print('popt',popt)
    
    plt.plot(t, spe_spectrum_gauss_exp(t, *popt), label = 'spectrum fit') #funzione che si ottiene con il fit  utilizzando i popt
    
    gauss_spe = lambda x : gauss(x, *popt[:3])  #gaussiana ottenuta considerando i primi 3 parametri ottimali
    nspe, spe_er = integ.quad(gauss_spe,0,t[-1])
    print('nspe',nspe)
    plt.plot(t, gauss(t, *popt[:3]),label='Gaussiana')
    
    
    gauss_bs = lambda x : expgauss(x, *popt[3:])  #esponenziale convoluta con gaussiana ottenuta considerando gli ultimi 3 parametri ottimali
    nbs, bs_er = integ.quad(gauss_bs,0,t[-1])
    print('nbs',nbs)
    plt.plot(t, expgauss(t, *popt[3:]),label=f'Expgauss - {nbs/(nspe+nbs)*100:.1f}% Non-Returning')
    
    
    spe_int = lambda x : spe_spectrum_gauss_exp(x, *popt)  #funzione ottenuta dalla somma della gaussiana e dell'esponenziale convoluta con una gaussiana
    plt.plot(t, spe_spectrum_gauss_exp(t, *guess), label = 'guess')
    
    #plt.title(f'ABALONE at {volts} kV - LED at {ledv:.1f} V')
    plt.xlabel('Scala lineare - area (ADC x 10ns)')
    plt.ylabel('Scala lineare - conteggi')
    plt.tick_params(axis='x',labelsize=12)
    plt.tick_params(axis='y',labelsize=12)
    plt.legend(fontsize=12)
    #plt.yscale('log')
    plt.ylim(1,1000)
    if save: plt.savefig('fit_Gauss_Exp_data.png',dpi=800)
    return popt[1], popt[2], nspe, nbs
```

```python
fit_spe_spectrum_gauss_exp(data['area'], volts = 24, low = 0, high = 10000, bins = 300,spe_div=1500)
```

```python
def spe_spectrum_gauss_landau(x,a1,mu1,sigma1,a2,loc,scale):
    return gauss(x,a1,mu1,sigma1)+ landau(x,a2,loc,scale)

def fit_spe_spectrum_gauss_landau(area, bins = 200, volts = 10, ledv = 3, low = 0, high = 10000, spe_div = 4000,save = True):
    area_space = np.linspace(low,high, bins)
    h, t = np.histogram(area, bins=area_space)
    plt.figure(figsize=(8,4.5))
    tc = (t[:-1] + t[1:])/2.
    a1 = plt.hist(area,bins=area_space,histtype='step',lw=2,density=False)
    
    idx1 = np.where(t>spe_div)[0][0]
    imax = np.argmax(h[idx1:])+idx1
    mu, hmax = t[imax], h[imax]
    idx = np.where(h[idx1:]>hmax/2) # fwhm 
    ilo, ihi = idx[0][0], idx[0][-1]
    sig = (t[ihi]-t[ilo]) / 2.355
    
    guess = (hmax, mu, sig, 200, 1150, 600)
    
    bounds = ([hmax*0.9, mu-sig, 0, 100, 1100, 150],
              [1.1*hmax, mu+sig, 2*sig, 300, 1250, 700])
    
    #fit 
    
    popt, pcov = curve_fit(spe_spectrum_gauss_landau, tc[idx1:], h[idx1:], p0 = guess, bounds = bounds)
    perr = np.sqrt(np.diag(pcov))
    print(popt)
    
    plt.plot(t, spe_spectrum_gauss_landau(t, *popt), label = 'spectrum fit')
    spe_int = lambda x : spe_spectrum_gauss_landau(x, *popt)
        
    gauss_spe = lambda x : gauss(x, *popt[:3])
    nspe, spe_er = integ.quad(gauss_spe,0,t[-1])
    print('nspe',nspe)
    plt.plot(t, gauss(t, *popt[:3]),label='Gaussiana')
    
    landau_int = lambda x : landau(x, *popt[3:])
    nbs, bs_er = integ.quad(landau_int,0,t[-1])
    print('nbs',nbs)
    plt.plot(t, landau(t, *popt[3:]),label=f'Landau - {nbs/(nspe+nbs)*100:.1f}% Non-Returning')
    
    plt.plot(t, spe_spectrum_gauss_landau(t, *guess), label = 'guess')
    
    #plt.title(f'ABALONE at {volts} kV - LED at {ledv:.1f} V')
    plt.xlabel('Scala lineare - area (ADC x 10ns)')
    plt.ylabel('Scala lineare - conteggi')
    plt.tick_params(axis='x',labelsize=12)
    plt.tick_params(axis='y',labelsize=12)
    plt.legend(fontsize=12)
    #plt.yscale('log')
    plt.ylim(1,1e3)
    if save: plt.savefig('fit_Gauss_Landau_data.png',dpi=800)
    return popt[1], popt[2], nspe, nbs
```

```python
def expo(x, a, tau):
    return a*np.exp(-x/tau)
def noise_spectrum(x,a0,tau0,a1,x1,sigma1,a2,x2,sigma2,a3,x3,sigma3,a4,x4,sigma4):
    return expo(x,a0,tau0)+gauss(x,a1,x1,sigma1)+gauss(x,a2,x2,sigma2)+gauss(x,a3,x3,sigma3)+gauss(x,a4,x4,sigma4)

def fit_spectrum_low(area, bins = 200, volts = 10, ledv = 3, low = 0, high = 750, spe_div = 150,save = False):
    
    area_space = np.linspace(low,high, bins)
    h, t = np.histogram(area, bins=area_space)
    plt.figure(figsize=(12,6))
    tc = (t[:-1] + t[1:])/2.
    a = plt.hist(area,bins=area_space,histtype='step',lw=2,density=False)
    
    idx1 = np.where(t>spe_div)[0][0]
    imax = np.argmax(h[:idx1])
    x1, hmax = t[imax], h[imax]
    
    guess = (  5000, 180,  5000, x1,   20, 2000, 150, 50, 500, 270, 50, 200, 400, 50)
    
    bounds = ([4800, 120, 4000, x1-10, 10,  500, 140, 30, 50,  250, 30, 10, 380, 30],
              [6000, 200, 6000, x1+10, 30, 3000, 160, 70,1000, 290, 70, 700, 410, 70])
    
    #fit
    popt, pcov = curve_fit(noise_spectrum, tc[:idx1], h[:idx1], p0 = guess, bounds = bounds)
    perr = np.sqrt(np.diag(pcov))
    #print(popt)
    plt.plot(t, noise_spectrum(t, *popt), label = 'spectrum fit')
    #expgauss_spe = lambda x : noise_spectrum(x, *popt[:4])
    #landau_int = lambda x : gauss(x, *popt[4:])
    plt.plot(t, expo(t, *popt[:2]),label=f'noise')
    plt.plot(t, gauss(t, *popt[2:5]),label=f'SiPM dark noise 1: {popt[3]:.1f} +/- {perr[3]:.1f} (ADC x 10ns)')
    plt.plot(t, gauss(t, *popt[5:8]),label=f'SiPM dark noise 2: {popt[6]:.1f} +/- {perr[6]:.1f} (ADC x 10ns)')
    plt.plot(t, gauss(t, *popt[8:11]),label=f'SiPM dark noise 3: {popt[9]:.1f} +/- {perr[9]:.1f} (ADC x 10ns)')
    plt.plot(t, gauss(t, *popt[11:]),label=f'SiPM dark noise 4: {popt[12]:.1f} +/- {perr[12]:.1f} (ADC x 10ns)')
    #plt.plot(t, expo(t, *guess[:2]),label=f'noise')
    #plt.plot(t, gauss(t, *guess[2:5]),label=f'SiPM dark noise 1')
    #plt.plot(t, gauss(t, *guess[5:8]),label=f'SiPM dark noise 2')
    #plt.plot(t, gauss(t, *guess[8:11]),label=f'SiPM dark noise 3')
    #plt.plot(t, noise_spectrum(t, *guess), label = 'guess')
    #plt.title(f'ABALONE at {volts} kV - LED at {ledv:.1f} V')
    plt.xlabel('Scala lineare - area (ADC x 10ns)',fontsize=12)
    plt.ylabel('Scala lineare - conteggi',fontsize=12)
    plt.tick_params(axis='x',labelsize=12)
    plt.tick_params(axis='y',labelsize=12)
    plt.legend(fontsize=12)
    plt.ylim(1e1,1e4)
    #plt.xlim(-10,200)
    plt.yscale('log')
    if save: plt.savefig('fit_Gauss_Landau_data.png',dpi=800)
    print(popt)
    return popt
```

```python
parlow = fit_spectrum_low(data['area'], low = 0, high = 750, bins = 300, spe_div=500)
```

```python
def expgauss(x,a,mu,sigma,tau):
    return a/2 * np.exp((sigma**2-2*x*tau)/(2*tau**2))*(1-scipy.special.erf((sigma**2+tau*(mu-x))/(np.sqrt(2)*sigma*tau) ))
def spe_spectrum(x,a1,mu1,sigma1,a2,mu2,sigma2,tau2):
    return gauss(x,a1,mu1,sigma1)+ expgauss(x,a2,mu2,sigma2,tau2)
def all_fit(x,a1,mu1,sigma1,a2,mu2,sigma2,tau2,a0,tau0,a3,x3,sigma3,a4,x4,sigma4,a5,x5,sigma5,a6,x6,sigma6):
    return spe_spectrum(x,a1,mu1,sigma1,a2,mu2,sigma2,tau2)+noise_spectrum(x,a0,tau0,a3,x3,sigma3,a4,x4,sigma4,a5,x5,sigma5,a6,x6,sigma6)

def fit_spe_spectrum(area, parlow, bins = 200, low = 0, high = 100, spe_div = 40,save = False):
    area_space = np.linspace(low,high, bins)
    h, t = np.histogram(area, bins=area_space)
    #plt.figure(figsize=(8,4.5))
    plt.figure(figsize=(12,6))
    tc = (t[:-1] + t[1:])/2.
    a1 = plt.hist(area,bins=area_space,histtype='step',lw=2,density=False)
    #SPE guess
    #idx1, idx2 = np.where(t>fit_range[0])[0][0], np.where(t>fit_range[1])[0][0]
    idx1 = np.where(t>2000)[0][0]
    imax = np.argmax(h[idx1:])+idx1
    mu, hmax = t[imax], h[imax]
    idx = np.where(h[idx1:]>hmax/2) # fwhm 
    ilo, ihi = idx[0][0], idx[0][-1]
    sig = (t[ihi]-t[ilo]) / 2.355
    #BS guess
    #imax = np.argmax(h[:idx1])
    #mu0, hmax0 = t[imax], h[imax]
    #idx = np.where(h[:idx1]>hmax/2) 
    #ilo, ihi = idx[0][0], idx[0][-1]
    #sig0 = (t[ihi]-t[ilo]) / 2.355
    #guess = (hmax, mu, sig, hmax0, mu0, sig0, hmax/2, 20, 5)
    guess = (hmax, mu, sig, 100, 1000, 400, 1500)
    bounds = ([hmax*0.9, mu-sig, sig*0.8, 50, 100, 100, 600],
              [1.1*hmax, mu+sig, sig*1.2, 4000, 1500, 500, 5000])
    #fit 
    #a0,tau0, a1,x1,sigma1, a2,x2,sigma2, a3,x3,sigma3, a4,x4,sigma4 = parlow
    fit_high = lambda x,a1,mu1,sigma1,a2,mu2,sigma2,tau2 : all_fit(x,a1,mu1,sigma1,a2,mu2,sigma2,tau2,*parlow)
    
    popt, pcov = curve_fit(fit_high, tc[500:], h[500:], p0 = guess, bounds = bounds)
    perr = np.sqrt(np.diag(pcov))
    plt.plot(t, fit_high(t, *popt), label = 'spectrum fit')
    gauss_spe = lambda x : gauss(x, *popt[:3])
    expgauss_int = lambda x : expgauss(x, *popt[3:])
    spe_int = lambda x : fit_high(x, *popt)
    nspe, spe_er = integ.quad(gauss_spe,0,t[-1])
    #print('nspe',nspe)
    nbs, bs_er = integ.quad(expgauss_int,0,t[-1])
    #print('nbs',nbs)
    ntot, tot_er  = integ.quad(spe_int,0,t[-1])
    plt.plot(t, gauss(t, *popt[:3]),label=f'SPE at {popt[1]:.2f} $\pm$ {popt[2]:.2f} ADC x $\mu$s')
    plt.plot(t, expgauss(t, *popt[3:]),label=f'{nbs/(nspe+nbs)*100:.1f}% Non-Returning')
    plt.plot(t, noise_spectrum(t, *parlow), label = 'noise spectrum')
    
    plt.plot(t, fit_high(t, *guess), label = 'guess')
    #plt.plot(t, gauss(t, *guess[:3]),label=f'SPE')
    #plt.plot(t, expgauss(t, *guess[3:]),label=f'Non-Returning')
    plt.xlabel('area (ADC x $\mu$s)',ha='right',x=1,fontsize=12)
    plt.ylabel('counts',ha='right',y=1,fontsize=12)
    plt.tick_params(axis='x',labelsize=12)
    plt.tick_params(axis='y',labelsize=12)
    plt.legend(fontsize=12)
    plt.yscale('log')
    plt.ylim(1,1e4)
    if save: plt.savefig('plots/SPEfit.png',dpi=800)
    return popt#, nspe, nbs
```
