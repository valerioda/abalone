```python
def calculate_integrals( data, nn = 0, plot = False, save = False):
    if plot: plt.figure(figsize=(12,6))
    if nn == 0: nn = len(data)
    peakint = np.zeros(nn)
    print('Total events:',nn)
    t_start = time.time()
    for i in range(nn):
        listpeaks = prsu.search_peaks(data[i], 4, 10, False)
        try: peakint[i] = prsu.integral_central_peak(data[i],listpeaks, dtl=-10, dtr=3, tfit=30,
                                  tlim=200, tc = 6, tll = 8, tlr = 10, plot = plot)
        except: pass
        diff = time.time() - t_start
        if (i % 1000) == 0:
            print(f'event n. {i} area: {peakint[i]:.2f}, time to process: {diff:.2f}')
    if save: np.save('peakint_SiPM1_ABALONE_19kV.npy', peakint)
    return peakint
```

```python
#peakint = np.load('peakint_SiPM1_27_5_LED2p75.npy')
peakint = calculate_integrals( new_wf, nn = 0 )
```
