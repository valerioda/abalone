```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.integrate import quad
import scipy.special as sc
from scipy.integrate import odeint
import process_utility as psu
```

```python
data = psu.read_file('data/2020_11_09/SiPM1_27_5_LED2p75.dat')
```

```python
plt.figure(figsize=(12,6))
dts = np.arange(0,data.shape[1]*10,10)
psu.plot_waveform(dts,data[0])
ddt, ddata = psu.derive(dts, data[0])
sdt, sdata = psu.smooth(dts, data[0],5)
psu.plot_waveform(sdt,sdata)
```

```python
for i in range(5): psu.integral(dts,data[i],-50, 0, 500, 2000, 0, 1)
```

```python
nn = len(data)
peakint = np.zeros(nn)
print('Total events:',nn)
for i in range(nn):
    peakint[i] = psu.integral(dts,data[i],-50, 0, 500, 2000)
    if (i % 1000) == 0: print('event n.',i,'area:',peakint[i])
```

```python
psu.histo_plot(peakint,10,8000,1000)
```
