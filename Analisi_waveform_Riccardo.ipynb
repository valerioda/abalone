```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import quad
import scipy.special as sc
from scipy.integrate import odeint
```

```python
def plotwfs(filename, events = 10, samples = 1024):
    data = np.fromfile(filename,  dtype=np.int16)
    n = len(data)/samples
    print(filename,'Number of events',n)
    data = np.array(np.array_split(np.array(data),n))
    
    # plot of waveforms
    fig, axs = plt.subplots(events, figsize=(12,10))
  #  fig.figure(figsize=(12,6))

    for i in range(events):
        axs[i].plot(data[i][:])
        axs[i].set(xlabel='samples', ylabel='channels')
        
def plotwfC(filename0, filename1, events = 10, samples = 1024):
    
    data0 = np.fromfile(filename0,  dtype=np.int16)
    data1 = np.fromfile(filename1,  dtype=np.int16)
    n0 = len(data0)/samples
    n1 = len(data1)/samples
    print(filename0,'Number of events',n0)
    print(filename1,'Number of events',n1)

    data0 = np.array(np.array_split(np.array(data0),n0))
    data1 = np.array(np.array_split(np.array(data1),n1))

    
    # plot of waveforms
    fig, axs = plt.subplots(events,2,  figsize=(15,2*events) )

    for i in range(events):
        axs[i,0].plot(data0[i][:], color='red')
        axs[i,0].set(xlabel='samples', ylabel='channels')
        axs[i,1].plot(data1[i][:], color='blue')
        axs[i,1].set(xlabel='samples', ylabel='channels')
        

        


        

```
