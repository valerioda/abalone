```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import quad
import scipy.special as sc
from scipy.integrate import odeint
```

```python
def plotwf(filename, events = 10, samples = 1024):
    data = np.fromfile(filename,  dtype=np.int16)
    n = len(data)/samples
    print(filename,'Number of events',n)
    data = np.array(np.array_split(np.array(data),n))
    
    # plot of waveforms
    plt.figure(figsize=(12,6))
    for i in range(events):
        plt.plot(data[i][:])
    #plt.plot(X[0,:], c='r', label='SiPM - LED waveform')
    plt.xlabel('samples')
    plt.ylabel('channels')
    #plt.legend()
```

```python
plotwf( f'/home/riccardo/Scrivania/Data/2020_10_28/SiPM1_27_0.dat', events =1)
```
