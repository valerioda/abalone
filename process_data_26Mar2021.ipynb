# SiPM2 at 32 V & LED at 2.81 V (threshold 15)

```python
data = psu.read_file('/data/abalone/2021_03_26/SiPM2_t15_LED2p81.dat')
```

```python
for i in range(5):
    idx = int(len(data)*np.random.rand())
    listpeaks = psu.search_peaks(data[idx], 4, 4, False)
    psu.integral_peaks(data[idx],listpeaks,-4,1,10,100,5,8,10,plot=True)
```

```python
def histo_plot(file,xrange=(0,20),bins=500,log=False):
    peakint = np.load(file)
    peakint = np.concatenate(peakint)
    peakint = peakint[peakint != np.array(None)]
    print('Number of integrals',len(peakint))
    hh, xx = np.histogram(peakint, bins=bins, range=(xrange[0],xrange[1]))
    plt.figure(figsize=(12,6))
    plt.plot(xx[1:],hh,'-')
    plt.xlabel('area',ha='right',x=1)
    plt.ylabel('number of events',ha='right',y=1)
    if log: plt.yscale('log')
```

```python
# fitting only the central peak
histo_plot(file='2021_03_26/peakint_SiPM2_t15_LED2p81.npy',xrange=(0,12),bins=500,log=0)
```

```python
# fitting all peaks
histo_plot(file='2021_03_26/peakint_SiPM2_t15_LED2p81.npy',xrange=(0,12),bins=500,log=0)
```
