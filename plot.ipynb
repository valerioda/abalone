```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import quad
import scipy.special as sc
from scipy.integrate import odeint

X, Y = [], []
events = 10
samples = 1024

for idx, line in enumerate(open('wave0.txt', 'r')):
    if idx < events*samples:
        X.append(float(line))
    else: break

data = np.array(np.array_split(np.array(X),events))

# plot of waveforms
plt.figure(figsize=(12,6))
for i in range(events):
    plt.plot(data[i][:])
#plt.plot(X[0,:], c='r', label='SiPM - LED waveform')
plt.xlabel('samples')
plt.ylabel('channels')
#plt.legend()
```
